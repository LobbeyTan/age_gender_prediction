```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.utils import shuffle
```

```python
df = pd.read_csv("./wiki/preprocessed_data.csv")
```

```python
df.shape
```

### Randomly Plot Images

```python
from random import randint

for _ in range(10):
    i = randint(0, df.shape[0])
    
    img = plt.imread(f"./wiki/{df['image_path'][i]}")
    print(f"path: ./wiki/{df['image_path'][i]} | size: {img.size}")
    print(f"Age: {df['age'][i]} | Gender: {df['gender'][i]}")
    
    plt.figure(i)
    plt.imshow(img)
    plt.axis("off")
    plt.plot()
    plt.show()
```

### Males vs Females

```python
plt.bar(x=["female", "male"], height=[(df['gender'] == 0).sum(), (df['gender'] == 1).sum()],  color=["blue", "red"])
plt.show() 
```

```python
female_records = df[df['gender'] == 0]
male_records = shuffle( df[df['gender'] == 1]).reset_index()
```

```python
balance_dataset = shuffle(pd.concat([female_records, male_records.sample(female_records.shape[0])]))
```

```python
plt.bar(x=["female", "male"], height=[(balance_dataset['gender'] == 0).sum(), (balance_dataset['gender'] == 1).sum()],  color=["blue", "red"])
plt.show() 
```

```python
balance_dataset.to_csv("./wiki/balance_data.csv", index=False)
```
