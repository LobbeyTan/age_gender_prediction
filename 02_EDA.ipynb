```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("preprocessed_data.csv")
```

```python
df.shape
```

### Randomly Plot Images

```python
from random import randint

for _ in range(10):
    i = randint(0, 51837)
    
    img = plt.imread(f"./wiki/{df['image_path'][i]}")
    print(f"path: ./wiki/{df['image_path'][i]} | size: {img.size}")
    print(f"Age: {df['age'][i]} | Gender: {df['gender'][i]}")
    
    plt.figure(i)
    plt.imshow(img)
    plt.axis("off")
    plt.plot()
    plt.show()
```

### Males vs Females

```python
plt.bar(x=["female", "male"], height=[(df['gender'] == 0).sum(), (df['gender'] == 1).sum()],  color=["blue", "red"])
plt.show()
```

### Age Frequency Count

```python
plt.hist(df['age'], bins=10)
plt.show()
```
